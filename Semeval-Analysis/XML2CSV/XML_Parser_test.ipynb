```python
import itertools
from xml.etree import ElementTree
import csv

tree = ElementTree.parse('reviews.xml')
reviews_data = tree.getroot()
```

```python
reviews=[]
data = open('reviews.csv','w',encoding='utf-8')

# create the csv writer object

csvwriter = csv.writer(data)
count = 0


for review in reviews_data:
    if count == 0:
        reviews.append('RID')
        reviews.append('SID')
        reviews.append('text')
        reviews.append('polarity')
        reviews.append('category')
        reviews.append('target')
        csvwriter.writerow(reviews)
        reviews = []
        count=1
    
    sentences = review.find('sentences')
    
    RID = review.attrib['rid']
    for sentence in sentences:
        opinions = sentence.find('Opinions')
        reviews.append(RID)
#         if sentence!=None:
        SID = sentence.attrib['id']
        reviews.append(SID)
        text = sentence.find('text').text
        reviews.append(text)
    #     else:
    #         SID = 'NULL'
    #         reviews.append(SID)
    #         text = 'NULL'
    #         reviews.append(text)
        if opinions!=None:
            for opinion in opinions:
#                 if opinion!=None:
                polarity = opinion.attrib['polarity']
                reviews.append(polarity)
                category = opinion.attrib['category']
                reviews.append(category)
                target = opinion.attrib['target']
                reviews.append(target)
    #             else:
    #                 polarity="NULL"
    #                 category="NULL"
    #                 reviews.append(polarity)
    #                 reviews.append(category)

                csvwriter.writerow(reviews)
                reviews.remove(polarity)
                reviews.remove(category)
                reviews.remove(target)
        #csvwriter.writerow(reviews)
        reviews = []
data.close()
```
